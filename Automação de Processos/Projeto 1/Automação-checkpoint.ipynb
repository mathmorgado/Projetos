## Projeto Automação de processos
---
Projeto onde, através de códigos com python, iniciará o navegador e entrará no site onde está disponivel a planilha, assim efetuando os calculos necessários e enviando um email com os dados requeridos em um texto padronizado!

## Passos
* Passo 1: Abrir o Navegador.
* Passo 2: Abrir site que disponibiliza a planilha.
* Passo 3: Fazer o download da planilha.
* Passo 4: Fazer os cálculos necessários.
* Passo 5: Abrir Email.
* Passo 6: Enviar um email com os dados necessários em um texto padronizado.

```python
import pyautogui as pa
import pyperclip as pc
import pandas as pd
import time

time.sleep(10)

# Passo 1
pa.press("win")
pa.write("chrome")
pa.press("enter")

time.sleep(2.5)

# Passo 2
pa.hotkey('ctrl', 't') # Nova aba
pc.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga") # link do site
pa.hotkey('ctrl', 'v')
pa.press('enter')

time.sleep(3)

pa.click(x=400, y=288, clicks=2) # área de exportação

time.sleep(1.5)

# Passo 3
pa.click(x=400, y=288) # arquivo
time.sleep(1.5)
pa.click(x=1712, y=195, clicks=3) # tres pontinho
time.sleep(1)
pa.click(x=1545, y=586) # download

time.sleep(4)

pa.press('enter') # Confirmar download

time.sleep(1)

# Passo 4
tabela = pd.read_excel(r'C:\Users\Matheus\Downloads\Vendas - Dez.xlsx')
faturamento = tabela['Valor Final'].sum()
qtde_prod = tabela['Quantidade'].sum()

# Passo 5
pa.press("win")
pa.write("email")
pa.press("enter")

time.sleep(3)

pa.click(x=118, y=106) # Novo email

time.sleep(0.5)

# Passo 6
pa.write('matheusrmorgado@hotmail.com')
pa.press('tab')
pa.press('tab')
pa.press('tab')
pc.copy('Relatório de Vendas')
pa.hotkey('ctrl', 'v')
time.sleep(0.25)
pa.press('tab')
time.sleep(0.25)
texto = f"""
Prezados, bom dia!

O faturamento do mês de dezembro foi: R$ {faturamento:,.2f}
A quantidade de produtos vendidos foi: R$ {qtde_prod:,}

abs
Matheus Morgado
"""
pc.copy(texto)
pa.hotkey('ctrl', 'v')
pa.hotkey('ctrl', 'enter')
```
